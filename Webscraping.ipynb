```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
link = "https://understat.com/player/7781"
res = requests.get(link)
soup = BeautifulSoup(res.content,'lxml')
scripts = soup.find_all('script')

# Get the grouped stats data

strings = scripts[1].string
```

```python
# Clean the data and remove the unncecessary clutter

ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')
data = json.loads(json_data)
```

```python
df = pd.DataFrame(data)
```

```python
# Since it's a dictionary of dictionaries (Nested dictionaries), we must call several keys and create DataFrames of them

df = pd.DataFrame(data['season'])
df
```

```python
df = pd.DataFrame(data['shotTypes'])
df
```

```python
# To unnest them what we can do is to apply the series function and create more dataframes out of it

df['2019'].apply(pd.Series)
```
